```python
import datetime
```

```python
month_arr = []
year_arr = []
for date in weather_df['"Date"']:
    date = date.replace('"', "")
    date = pd.to_datetime(date, format="%m/%d/%Y")
    month_arr.append(date.month)
    year_arr.append(date.year)
    
month_col = pd.Series(month_arr, name='"month"')
year_col = pd.Series(year_arr, name='"year"')
weather_df["year"] = year_col
weather_df["month"] = month_col

print(weather_df.head())
```

```python
group_year = weather_df.groupby(['year' , 'month']).mean()
group_year['"Precipitation_In"']
```

```python
group_year = weather_df.groupby(['year' , 'month']).median()
group_year['"Precipitation_In"']
```
